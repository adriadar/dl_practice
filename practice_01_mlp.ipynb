````python
def set_random_seed(seed):
    torch.backends.cudnn.deterministic = True
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
````

Вы можете придерживаться любой адекватной струкуры кода, но мы советуем воспользоваться следующими сигнатурами функций.
````python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader

from tqdm.notebook import tqdm
from IPython.display import display, clear_output
````

````python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
````

````python
ss2 = StandardScaler()
y_train = ss2.fit_transform(y_train[:, None]).reshape(-1)
y_test = ss2.transform(y_test[:, None]).reshape(-1)
````

````python
set_random_seed(101)
````

````python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2)
````

````python
class myDataset(Dataset):
    def __init__(self, X, y):
        super().__init__()
        self.X = X
        self.y = y
        self.len = len(X)

    def __len__(self):
        return self.len

    def __getitem__(self, index):
        return self.X[index], self.y[index]
````

````python
BATCH_SIZE = 128

train_set = myDataset(X_train, y_train)
train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)

val_set = myDataset(X_val, y_val)
val_loader = DataLoader(val_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)

test_set = myDataset(X_test, y_test)
test_loader = DataLoader(test_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
dls = {'train': train_loader, 'val': val_loader, 'test': test_loader}
````

````python
class Model(nn.Module):
    def __init__(self, in_features = 90, out_features = 1):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.hidden_size = int(np.sqrt(in_features + out_features))

        self.sequential = nn.Sequential(
            nn.Linear(90, 130),
            nn.LeakyReLU(),
            nn.Linear(130, 80),
            nn.LeakyReLU(),
            nn.Linear(80, 40),
            nn.LeakyReLU(),
            nn.Linear(40, 1),
        )

    def forward(self, x):
        x = self.sequential(x)
        return x
````

````python
LR = 3e-3
WEIGHT_DECAY = 4e-3
DEVICE = 'cpu'
GAMMA = 0.9995


model = Model()
model.to(DEVICE)

loss_fn = nn.MSELoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=GAMMA)
````

````python
metrics_dict = {
    "Epoch": [],
    "Train_RMSE": [],
    "Test_RMSE": [],
}

# Train loop
for epoch in tqdm(range(8)):
    metrics_dict["Epoch"].append(epoch)
    for stage in ['train', 'test']:
        with torch.set_grad_enabled(stage == 'train'):
            if stage == 'train':
                model.train()
            else:
                model.eval()

            loss_at_stage = 0 
            for batch in dls[stage]:
                x_batch, y_batch = batch
                x_batch, y_batch = x_batch.to(DEVICE), y_batch.to(DEVICE)

                y_pred = model(x_batch.float()).view(-1)
                loss = loss_fn(y_pred, y_batch.float())
                if stage == "train":
                    loss.backward()
                    optimizer.step()
                    scheduler.step()
                    optimizer.zero_grad()
                with torch.no_grad():
                    loss_at_stage += (torch.square((y_pred - y_batch) * ss2.scale_).sum()).item()
            rmse_at_stage = (loss_at_stage / len(dls[stage].dataset)) ** (1/2)
            metrics_dict[f"{stage.title()}_RMSE"].append(rmse_at_stage)
            
    clear_output(wait=True)
    display(pd.DataFrame(metrics_dict))
````

````python
res = pd.DataFrame(metrics_dict)

import matplotlib.pyplot as plt

plt.plot(res['Epoch'],res['Train_RMSE'] , label = "Train RMSE")
plt.plot(res['Epoch'],res['Test_RMSE'] , label = "Test RMSE")

plt.xlabel('epoch')
plt.ylabel('rmse')
plt.title('RMSE')

plt.legend()
plt.show()
````

````python
def rmse(preds, y_true):
    return ((preds - y_true) ** 2).mean() ** (1/2)

def eval_loop(model, X_val, X_test, y_val, y_test, model_is_nn=True, model_name="final"):
    print(f"Model {model_name}")
    for name, x_eval, y_eval in zip(['val', 'test'], [X_val, X_test], [y_val, y_test]):
        if model_is_nn:
            with torch.no_grad():
                preds = model(torch.Tensor(x_eval)).view(-1)
        else:
            model.fit(X_train, y_train)
            preds = model.predict(x_eval)
        preds_rescheduled = preds * ss2.scale_ + ss2.mean_
        y_eval_rescheduled = y_eval * ss2.scale_ + ss2.mean_
        rmse_val = rmse(preds_rescheduled, y_eval_rescheduled)
        print(f'{name}: RMSE = {rmse_val:.2f}')

eval_loop(model, X_val, X_test, y_val, y_test, model_is_nn=True, model_name="final")
````
